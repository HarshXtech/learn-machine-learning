```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
```

```python
# Preprocessing the data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
df_encoded
```

```python
x = df_encoded.drop('charges', axis="columns")
y = df['charges']
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
```

```python
models = {
    'Linear' : LinearRegression(),
    'Ridge' : Ridge(),
    'lasso' : Lasso(),
    'Elastic' : ElasticNet()
}
models.values()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
result = {}
plt.figure(figsize=(12, 8))
for name, model in models.items():
    model.fit(x_scaled, y)

    y_pred = model.predict(x_scaled)

    result[name] = {
        'name' : name,
        'mse' : mean_squared_error(y, y_pred),
        'mae' : mean_absolute_error(y, y_pred),
    }

    plt.scatter(y, y_pred, label=name, alpha=0.4)

minmum = min(min(y), min([min(model.predict(x_scaled)) for model in models.values()]))
maxmum = max(max(y), max([max(model.predict(x_scaled)) for model in models.values()]))
plt.plot([minmum, maxmum], [minmum, maxmum,] ,'k--' )
plt.tight_layout()
plt.legend()
plt.grid(True)
plt.show()
```
