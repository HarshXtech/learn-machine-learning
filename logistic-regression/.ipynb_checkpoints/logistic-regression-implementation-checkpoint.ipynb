```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# create database
```

```python
x, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=42)
```

```python
model = LogisticRegression()
model
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
# prabability of data where it will fall into. 0 or 1?
# model.predict_proba(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# 118 : true positive
# 17 : false positive (actual value : 0 , predicted value : 1)
# 29: false negative (actual value : 1, prediected value : 0)
# 136 : true negative
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

### Hyperparameter tuning and cross validation

```python
model
```

```python
panelty = ['l1', 'l2', 'elasticnet']
c_values = [100, 10, 1.0, 0.1, 0.01]
solver = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
```

```python
params = dict(penalty=panelty, C = c_values, solver=solver)
```

```python
### Grid Search CV : finding best parameter that will fit for this dataset
```

```python
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model, param_grid=params, scoring="accuracy", cv=cv, n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train, y_train)
```

```python
 grid.best_params_
```

```python
grid.best_score_
```

```python
grid.predict_proba(x_test)
```

```python
y_pred = grid.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

#### Randomized Search CV

```python
randomcv = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy')
```

```python
randomcv.fit(x_train, y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred = randomcv.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
