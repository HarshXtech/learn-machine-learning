```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'rating' : [8.0, 6.2, 7.2, 8.2],
    'duration' : [160, 170, 168, 155],
    'genre' : ['action', 'action', 'comedy', 'comedy']
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=1)

x = df[['rating', 'duration']]
y = df['genre']
model.fit(x, y)
```

```python
new_point = [[7.4, 114]]

print(model.predict(new_point))
print(model.kneighbors(new_point))
```

```python
plt.scatter(df['rating'], df['duration'])
plt.scatter(7.4, 114, c='r')
plt.plot([new_point[0][0], df.iloc[3]['rating']], [new_point[0][1], df.iloc[3]['duration']])
```
