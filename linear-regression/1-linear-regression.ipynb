```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1,2,3,4,5]).reshape(-1, 1)
y = np.array([40, 50, 60, 65, 75])
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
y_pred
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred)
```

```python
model.predict([[6]])
```
