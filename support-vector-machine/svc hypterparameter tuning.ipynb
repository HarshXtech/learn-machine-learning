```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
### Create a Synthetic Datapoints!
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0)
```

```python
pd.DataFrame(x)[0]
```

```python
# Scatterplot: to visualize the features. 

sns.scatterplot(x=pd.DataFrame(x)[0], y=pd.DataFrame(x)[1], hue=y)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
svc = SVC(kernel='linear')

svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)

print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(x_train, y_train)

y_pred = rbf.predict(x_test)

print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
polynomial = SVC(kernel='poly')

polynomial.fit(x_train, y_train)

y_pred = polynomial.predict(x_test)

print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sigmoid = SVC(kernel="sigmoid")

sigmoid.fit(x_train, y_train)

y_pred = sigmoid.predict(x_test)


print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
params = {
    'C' : [0.1, 1, 10, 100, 1000],
    'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel' : ['rbf']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=SVC(),  param_grid=params, cv=5, refit=True, verbose=4)
```

```python
grid.fit(x_train, y_train)

y_pred = grid.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
