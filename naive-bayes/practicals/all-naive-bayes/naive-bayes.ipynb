```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('website-data.csv')
```

```python
df.drop('Unnamed: 0', axis='columns', inplace=True)
```

```python
df
```

## Gaussian distribution

```python
from sklearn.naive_bayes import GaussianNB
```

```python
x = df[['age', 'monthly_usage_hours']]
```

```python
y = df['churn']
```

```python
model = GaussianNB()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y, y_pred)
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
x = df[['subscribed_newsletter', 'support_calls']]
```

```python
model = BernoulliNB()
model.fit(x, y)

y_pred = model.predict(x)
```

```python
accuracy_score(y, y_pred)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
x = df[['support_calls', 'website_visits']]
```

```python
model = MultinomialNB()
model.fit(x, y)

y_pred = model.predict(x)
```

```python
accuracy_score(y, y_pred)
```
