```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset.keys()
```

```python
print(dataset.DESCR)
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df
```

```python
dataset.data
```

### Will try to reduce the number of features. will select limited number of features

```python
# standardization
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()

# will get the mean and standar daviation of the data.
scalar.fit(df)
```

```python
# will use that mean and std to transform data
scaled_data = scalar.transform(df)
```

```python
scaled_data
```

```python
## Apply PCA algorithm
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
data = pca.fit_transform(scaled_data)
```

```python
data
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(data[:, 0], data[:, 1], c=dataset['target'], cmap='plasma')
plt.xlabel('first feature')
plt.ylabel('second feature')
```
